```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ref = np.loadtxt('examples/0717-up.dat');
scale = 32768;
ref = np.round(scale * ref);
plt.plot(ref)
```

```python
y = np.fromfile('rx.raw', np.dtype('<i2'));
u0 = np.fromfile('u0.raw', np.dtype('<i2'));
u1 = np.fromfile('u1.raw', np.dtype('<i2'));
u2 = np.fromfile('u2.raw', np.dtype('<i2'));
u3 = np.fromfile('u3.raw', np.dtype('<i2'));
u4 = np.fromfile('u4.raw', np.dtype('<i2'));
Fs = 62500;
```

```python
plt.figure(figsize=(15,15));
plt.plot(y[:10000]);
```

```python
xc0 = np.correlate(u0,ref,"full");
xc1 = np.correlate(u1,ref,"full");
xc2 = np.correlate(u2,ref,"full");
xc3 = np.correlate(u3,ref,"full");
xc4 = np.correlate(u4,ref,"full");
#xr = np.correlate(ref,ref,"full")

plt.figure(figsize=(15,15));
#start = 868; end = 1180
start = 980; end = 1100
#start = 500; end = 11200
#start = 0; end = 11200
plt.plot(xc0[start:end], label='ch1')
plt.plot(xc1[start:end], label='ch2')
plt.plot(xc2[start:end], label='ch3')
plt.plot(xc3[start:end], label='ch4')
plt.plot(xc4[start:end], label='ch5')
#plt.plot(xr[start:end], label='autocorr')
plt.legend(loc='upper right')
```
